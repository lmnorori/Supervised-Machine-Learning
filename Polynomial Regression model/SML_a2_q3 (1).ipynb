### Exercise 3a:

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
random.seed(0)
import turicreate as tc
```

### Exercise 3b: 

>**Create an array of coefficients named _coefs_ to define a new polynomial 
where:**  


$f(x) = (a + 1)x^3 + (b + 1)x^2 + (c + 1)x + d $  
* a is the last digit of your student number
* b is the second to last
* c is the third last digit 
* d is the fourth last digit

```python
coefs = [5,2,5,5]

def polynomial(coefs, x):
    n = len(coefs)
    return sum([coefs[i]*x**i for i in range(n)])

def draw_polynomial(coefs):
    n = len(coefs)
    x = np.linspace(-5, 5, 1000)
    plt.ylim(-20,20)
    plt.plot(x, sum([coefs[i]*x**i for i in range(n)]), linestyle='-', color='black')

draw_polynomial(coefs)
```

```python
X = []
Y = []
for i in range(40):
    x = random.uniform(-5,5)
    y = polynomial(coefs, x) + random.gauss(0,2)
    X.append(x)
    Y.append(y)
```

```python
plt.scatter(X, Y)
#draw_polynomial(coefs)
```

```python
data = tc.SFrame({'x':X, 'y':Y})
data
```

```python
for i in range(2,200):
    string = 'x^'+str(i)
    data[string] = data['x'].apply(lambda x:x**i)
data
```

### Exercise 3c:

>**Split the data so that 75% is used for training**

```python
train, test = data.random_split(.75, seed=0)
```

### Exercise 3d:

>**Create a model called _model_elastic_reg_ that uses an L1 penalty of 0.1 and an L2 penalty of 0.1. (Using both L1 and L2 penalties is known as elastic regularization). Display the results.**

```python
def display_results(model):
    coefs = model.coefficients
    print("Training error (rmse):", model.evaluate(train)['rmse'])
    print("Testing error (rmse):", model.evaluate(test)['rmse'])
    plt.scatter(train['x'], train['y'], marker='o')
    plt.scatter(test['x'], test['y'], marker='^')
    draw_polynomial(coefs['value'])
    plt.show()
    print("Polynomial coefficients")
    print(coefs['name', 'value'])
```

```python
model_elastic_reg = tc.linear_regression.create(
    train, target='y', l1_penalty=0.1, l2_penalty=0.1, verbose=False, validation_set=None)
```

```python
display_results(model_elastic_reg)
```

### Exercise 3e:

>**Compare the training and testing error of _model_elastic_reg_. Comment on what this implies in terms of overfitting and underfitting.**

Based in the errors, the model is underfitting the data since both training and testing errors appear to be high. 

### Exercise 3f:

>**Find and display the predictions of _model_elastic_reg_ for each point in the testing data.**

```python
predictions = test['x', 'y']
predictions['Elastic reg'] = model_elastic_reg.predict(test)
```

```python
predictions
```

### Exercise 3g:

>**Create a new model named _optimal_model_. Experiment with different values for L1 and L2 penalty until you obtain better results than _model_elastic_reg_ for each point in the testing data. Display the results and make predictions for the test data.**

```python
optimal_model = tc.linear_regression.create(
    train, target='y', l1_penalty=0.0, l2_penalty=0.0001, verbose=False, validation_set=None)
```

```python
display_results(optimal_model)
```

```python
predictions = test['x', 'y']
predictions['Optimal reg'] = optimal_model.predict(test)
```
